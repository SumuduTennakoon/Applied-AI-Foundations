## 4. Training Model

### Build Simple Model

````python
from sklearn import linear_model
# https://scikit-learn.org/stable/modules/linear_model.html
````

````python
model = linear_model.LinearRegression()
print(model)
````

````python
X = ["curb-weight"]

y_actual = "price"
````

````python
model.fit(data[X], data[y_actual])
````

Drop Missing Value Rows

````python
data[y_actual].info()
````

````python
# Check missing values
data[data[y_actual].isnull()]
````

````python
# Keep records as demo dataset
demo_data = data[data[y_actual].isnull()]
````

````python
# Drop records with missing y_actual

data = data.dropna(subset=[y_actual])

data.info()
````

Retarin the model

````python
model.fit(data[X], data[y_actual])
````

````python
#Find model parameters
coefficients = model.coef_
intercept = model.intercept_

print(pd.DataFrame(data={'features':X, 'coefficients':coefficients}))
print('\n') # Add new line to print
print(F"Intercept = {intercept}")
````

## Regression Equation:

$$y = c + m*x $$

$$price = intercept + a_1*weight $$

$$price = -19553.57 +  12.818847 * weight  $$

### Predict

````python
test_weight = 2500.0

model.predict([[test_weight]])
````

### Test Run on Dataset

````python
data[y_predict] = model.predict(data[X])
data['difference'] = data[y_actual] - data[y_predict]
data[[y_actual, y_predict, 'difference']]
````

````python
import matplotlib.pyplot as plt

%matplotlib inline

# Plot predicted vs actual
plt.scatter(data[y_actual], data[y_predict])
plt.xlabel(y_actual)
plt.ylabel(y_predict)
plt.title('Car Price Predictions')

plt.show()
````

### Attempt to Train

````python
# List of Features
X = ['curb-weight', 'highway-mpg', 'city-mpg', 'horsepower', 'drive-wheels-num', 'num-of-doors']

# Target
y_actual = 'price'

# Label for Predicted Value Column
y_predict = 'predicted_price'
````

````python
model.fit(data[X], data[y_actual])
````

Dropping missing value rows

````python
data[X].info()
````

````python
# Check missing values
data[data['horsepower'].isnull()]
````

````python
# Check missing values
data[data['num-of-doors'].isnull()]
````

What are the other options?

### Attempt to Train

````python
model.fit(data[X], data[y_actual])
````

````python
#Find model parameters
coefficients = model.coef_
intercept = model.intercept_

print(pd.DataFrame(data={'features':X, 'coefficients':coefficients}))
print('\n') # Add new line to print
print(F"Intercept = {intercept}")
````

````python
data[y_predict] = model.predict(data[X])
data['difference'] = data[y_actual] - data[y_predict]
data[[y_actual, y_predict, 'difference']]
````

````python
import matplotlib.pyplot as plt

%matplotlib inline

# Plot predicted vs actual
plt.scatter(data[y_actual], data[y_predict])
plt.xlabel(y_actual)
plt.ylabel(y_predict)
plt.title('Car Price Predictions')

plt.show()
````

## 5. Evaluate Model

* Mean squared error: https://en.wikipedia.org/wiki/Mean_squared_error
* Coefficient of determination (R^2): https://en.wikipedia.org/wiki/Coefficient_of_determination


````python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(data[y_actual], data[y_predict])
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(data[y_actual], data[y_predict])
print("R2:", r2)
````

## 6. Recap

### Single Valriable

````python
X = ['curb-weight']

# Fit
model.fit(data[X], data[y_actual])

#Find model parameters
coefficients = model.coef_
intercept = model.intercept_

print(pd.DataFrame(data={'features':X, 'coefficients':coefficients}))
print('\n') # Add new line to print
print(F"Intercept = {intercept}")
data[y_predict] = model.predict(data[X])
data['difference'] = data[y_actual] - data[y_predict]
data[[y_actual, y_predict, 'difference']]

# Plot predicted vs actual
plt.scatter(data[y_actual], data[y_predict])
plt.xlabel(y_actual)
plt.ylabel(y_predict)
plt.title('Car Price Predictions')

plt.show()

# Evaluate
mse = mean_squared_error(data[y_actual], data[y_predict])
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(data[y_actual], data[y_predict])
print("R2:", r2)
````
